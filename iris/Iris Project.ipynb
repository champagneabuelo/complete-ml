## Data exploration and analysis

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)

%matplotlib inline
```

```python
df = pd.read_csv('iris.data', header=-1)
df.head()
```

```python
col_name = ["sepal length", "sepal width", "petal length", "petal width", "class"]
```

```python
df.columns = col_name
df.head()
```

## Iris Data from Seaborn

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
df.describe()
```

```python
iris.describe()
```

```python
print(iris.groupby('species').size())
```

## Visualization

```python
sns.pairplot(iris, hue='species', height=2.5, aspect=1)
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris)
```

```python
iris.boxplot(by='species', figsize=(12,8))
```

```python
pd.plotting.scatter_matrix(iris, figsize=(12,8))
plt.show()
```
